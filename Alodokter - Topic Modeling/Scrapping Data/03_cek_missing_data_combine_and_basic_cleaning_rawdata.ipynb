### **Clean rawdata**

```python
import re

data_cleaned = data.copy()

# Fungsi untuk membersihkan teks
def clean_text(text):
    # Hapus tag HTML
    text = re.sub(r'u003c.*?u003e', ' ', text)s
    # Hapus URL
    text = re.sub(r'https?://\S+', ' ', text)
    # Hapus karakter-karakter spesial lainnya yang tidak diinginkan
    text = re.sub(r'\\u003c|\\u003e|\\n|\\r|\\|"', ' ', text)
    return text.strip()

# Menerapkan fungsi clean_text pada kolom 'Question' dan 'Answer'
data_cleaned['Question'] = data_cleaned['Question'].apply(clean_text)
data_cleaned['Answer'] = data_cleaned['Answer'].apply(clean_text)
```

```python
# Mapping nama bulan dalam bahasa Indonesia ke format yang dapat dikenali Python
bulan_mapping = {
    'Januari': 'January',
    'Februari': 'February',
    'Maret': 'March',
    'April': 'April',
    'Mei': 'May',
    'Juni': 'June',
    'Juli': 'July',
    'Agustus': 'August',
    'September': 'September',
    'Oktober': 'October',
    'November': 'November',
    'Desember': 'December'
}

# Mengubah nama bulan dalam kolom 'Answer Date' sebelum konversi ke datetime
def convert_to_english_month(date_str):
    split_date = date_str.split()
    if len(split_date) >= 2 and split_date[1] in bulan_mapping:
        split_date[1] = bulan_mapping[split_date[1]]
    return ' '.join(split_date)

data_cleaned['Datetime'] = data_cleaned['Answer Date'].apply(convert_to_english_month)

# Mengubah kolom 'Answer Date' ke tipe datetime
data_cleaned['Datetime'] = pd.to_datetime(data_cleaned['Datetime'], format='%d %B %Y, %H:%M', errors='coerce')

# Membuat kolom baru 'Year' berisi tahun dari kolom 'Datetime'
data_cleaned['Year'] = data_cleaned['Datetime'].dt.year
```
